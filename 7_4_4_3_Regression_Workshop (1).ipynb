#Boston Regression

Using statsmodels and the `boston` dataset, make a regression model to predict house prices. Don't forget to add a constant (intercept) term. Note that statsmodels can take a `pd.DataFrame` as an input for `X`.

Report the $R^2$ and coefficients on each feature.

```
from sklearn.datasets import load_boston

boston = pd.DataFrame(load_boston().data,columns = load_boston().feature_names )
y = load_boston().target
```

```
# linear regression
X = sm.add_constant(boston)
sm.OLS(y, X).fit().summary()
```

```
# deleting non-related column
X = X.drop(columns=['INDUS', 'AGE'])
sm.OLS(y, X).fit().summary()
```

#Regressive Tennis T-Test

Use a linear regression and statsmodels to run a t-test on whether Federer scores more points than his opponents in the `tennis.csv` dataset.


Also answer the following: should your regression include a constant term? Why or why not? How would it change the interpretation of your coefficient and p-value?

```
from google.colab import files
uploaded = files.upload()
```

```
df = pd.read_csv('tennis.csv')
df.head()
```

```
df = df.dropna()
df['win'] = 0
df.loc[(df['winner']==df['player1 name']),['win']] = 1
```

```
sc.ttest_1samp(df['win'], 0.5)
```

By running a t-test with pvalue=2.6923903460915165e-116, null hypothesis is rejected and Federer scores more points than his opponents.

```
df.info()
```

```
# get string columns
obj_col = []
for column in df.columns:
  if df[column].dtype == 'object':
    obj_col.append(column)
```

```
df_r = df.drop(columns=obj_col)
df_r.info()
```

```
# regression without intercept
X = df_r.drop(columns=['win'])
y = df_r['win']

sm.OLS(y, X).fit().summary()
```

```
# regression with intercept
X = sm.add_constant(X)

sm.OLS(y, X).fit().summary()
```

Regression model without intercept performs better with higher R^2 value.

#College Admissions

Using the `college.csv` dataset, answer the following:

1. Do private schools see more admissions overall? Hint: use a binary explanatory variable for `Private`

2. Is the relation between `Top10perc` and `Top25perc` best fit using a model with only one variable, or one variable and a polynomial of degree 2? Is a constant term useful? How would you select for the best of these model specifications?

```
from google.colab import files
uploaded = files.upload()
```

```
import io
df = pd.read_csv(io.BytesIO(uploaded['college.csv']))
df = df.dropna()
```

```
df.head()
```

```
df_p = df.loc[:,['Private']]
df_p = df_p.dropna()
df_p.loc[df_p['Private']=='Yes',['Private']] = 1
df_p.loc[df_p['Private']!=1,['Private']] = 0
```

```
sc.ttest_1samp(df_p['Private'], 0.5)
```

As t-test result shown, null hypothesis can be rejected. Thus, private schools see more admissions overall.

```
X = df['Top10perc']
y = df['Top25perc']

# linear no intercept
sm.OLS(y, X).fit().summary()
```

```
# linear with intercept
X = sm.add_constant(X)
sm.OLS(y, X).fit().summary()
```

```
# polynomial no intercept
X = df.loc[:,['Top10perc']]
X = np.concatenate([X, (df['Top10perc']**2)[:, np.newaxis]], axis=1)
 
sm.OLS(y, X).fit().summary()
```

```
# polynomial with intercept
X = sm.add_constant(X)

sm.OLS(y, X).fit().summary()
```

Among four diffrent models, polynomial without intercept performs best with highest R^2 value(0.975).
